```python
# Extract the data of the winning team who opted to bat first
bat_first = ipl[ipl['win_by_runs']!=0]
bat_first.head()
```

```python
# Plot the histogram w.r. to Runs
plt.figure(figsize=(10,5))
plt.hist(bat_first['win_by_runs'],color='blue')
plt.title("Distribution of runs")
plt.xlabel("Runs")
plt.show()
```

- Teams who opted to bat first had won by 0 to 20 runs out of 100 matches

```python
# Find out number of wins with respect to each time after batting first
bat_first['winner'].value_counts()
```

```python
bat_first_teams = list(bat_first['winner'].value_counts()[0:3].keys())
bat_first_count = list(bat_first['winner'].value_counts()[0:3])
 
```

```python
# Barplot of top 3 teams with most wins after opting to bat first
plt.figure(figsize=(8,5))
plt.bar(bat_first_teams,bat_first_count,color=['blue','orange','green'] )
plt.title("Distribution of runs")
plt.xlabel("Runs")
plt.show()
```

```python
 list(bat_first['winner'].value_counts())
```

```python
# Plot Pie Chart
plt.figure(figsize=(20,20))
plt.pie(list(bat_first['winner'].value_counts()),labels= list(bat_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.title("Distribution of wins")
plt.show()
```
